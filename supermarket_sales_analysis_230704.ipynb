```python
data_feature = data.drop(columns = ['Store ID ', 'Store_Sales'], axis=1)
data_target = data['Store_Sales']
```

```python
data_feature.shape
```

```python
data_target.shape
```

```python
data_target = np.array(data_target).reshape(-1, 1)
data_target.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_feature, data_target, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2)
```

```python
print(X_train.shape, y_train.shape)
print(X_val.shape, y_val.shape)
print(X_test.shape, y_test.shape)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print('íšŒê·€ê³„ìˆ˜:', lr.coef_)
print('ì ˆíŽ¸:', lr.intercept_)
```

```python
# train set ì„±ëŠ¥í‰ê°€
y_train_predict = lr.predict(X_train)
rmse = np.sqrt(mean_squared_error(y_train, y_train_predict))
r2 = lr.score(X_train, y_train)

print('RMSE:', rmse)
print('R2 score:', r2)
```

```python
# validation set ì„±ëŠ¥í‰ê°€
y_val_predict = lr.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_val, y_val_predict))
r2 = lr.score(X_val, y_val)

print('RMSE:', rmse)
print('R2 score:', r2)
```

### Ridge & Lasso Regression

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet

ridge = Ridge(alpha=10)
neg_mean_scores = cross_val_score(ridge, data_feature, data_target, scoring='neg_mean_squared_error', cv=5)
rmse = np.sqrt(-1 * neg_mean_scores)
avg_rmse = np.mean(rmse)

print('RMSE:', rmse)
print('í‰ê·  RMSE:', avg_rmse)
```

```python
lasso = Lasso(alpha=10)
neg_mean_scores = cross_val_score(lasso, data_feature, data_target, scoring='neg_mean_squared_error', cv=5)
rmse = np.sqrt(-1 * neg_mean_scores)
avg_rmse = np.mean(rmse)

print('RMSE:', rmse)
print('í‰ê·  RMSE:', avg_rmse)
```

```python
elasticnet = ElasticNet(l1_ratio=0.2)
neg_mean_scores = cross_val_score(elasticnet, data_feature, data_target, scoring='neg_mean_squared_error', cv=5)
rmse = np.sqrt(-1 * neg_mean_scores)
avg_rmse = np.mean(rmse)

print('RMSE:', rmse)
print('í‰ê·  RMSE:', avg_rmse)
```
